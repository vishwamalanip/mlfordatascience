```python
from matplotlib import pyplot as plt 
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('cardata.csv')
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

# Car_Name

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Car_Name_en =LabelEncoder()
dataset['Car_Name']=Car_Name_en.fit_transform(dataset['Car_Name'])
dataset.head(5)
```

# Fuel_Type

```python
dataset['Fuel_Type'].unique()
```

```python
Fuel_Type_en =LabelEncoder()
dataset['Fuel_Type']=Fuel_Type_en.fit_transform(dataset['Fuel_Type'])
dataset.head(5)
```

# Seller_Type

```python
dataset['Seller_Type'].unique()
```

```python
Seller_Type_en =LabelEncoder()
dataset['Seller_Type']=Seller_Type_en.fit_transform(dataset['Seller_Type'])
dataset.head(5)
```

# Transmission

```python
dataset['Transmission'].unique()
```

```python
Transmission_en =LabelEncoder()
dataset['Transmission']=Transmission_en.fit_transform(dataset['Transmission'])
dataset.head(5)
```

```python
inputdata= dataset.iloc[:,:-1]
inputdata
```

```python
outputdata=dataset['Selling_Price']#or outputdata=dataset.iloc[:,8]
outputdata
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit_transform(inputdata)
```

```python
inputdata=pd.DataFrame(ss.fit_transform(inputdata),columns=inputdata.columns)
```

```python
inputdata.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputdata,outputdata,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
sns.heatmap(data=dataset.corr(),annot=True)
plt.show()
```

# linear

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100 , lr.score(x_test,y_test)*100
#near to 100 gives good accuracy
```

# Lasso

```python
lr1=Lasso(alpha=4)
lr1.fit(x_train,y_train)
lr1.score(x_train,y_train)*100 , lr1.score(x_test,y_test)*100
#near to 100 gives good accuracy
```

# Ridge

```python
lr2=Ridge(alpha=10) #change the alpha value if it give diff answer every time that means good algo to use otherwise dont
lr2.fit(x_train,y_train)
lr2.score(x_train,y_train)*100 , lr2.score(x_test,y_test)*100
```

# ElasticNet

```python
lr3=ElasticNet(alpha=5)
lr3.fit(x_train,y_train)
lr3.score(x_train,y_train)*100 , lr3.score(x_test,y_test)*100
#near to 100 gives good accuracy
```

# DecisionTreeRegressor

```python
dt=DecisionTreeRegressor(max_depth=10) #max_depth=None by default
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100 , dt.score(x_test,y_test)*100
#near to 100 gives good accuracy
```

```python
#to check best result
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
#errors should be near to 0
```

```python
mean_squared_error(y_test,dt.predict(x_test))
```

```python
mean_absolute_error(y_test,dt.predict(x_test))
```

# SVR

```python
sv=SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100 , sv.score(x_test,y_test)*100
```

# RandomForestRegressor

```python
rf=RandomForestRegressor(n_estimators=100)#can be btw 0 to 100 #by default its 100
rf.fit(x_train,y_train)
rf.score(x_train,y_train)*100 , rf.score(x_test,y_test)*100
#near to 100 gives good accuracy
```

```python
mean_squared_error(y_test,rf.predict(x_test)),mean_absolute_error(y_test,rf.predict(x_test)) #which is less than decision tree
```

# KNeighborsRegressor

```python
knn=KNeighborsRegressor(n_neighbors=10)
knn.fit(x_train,y_train)
knn.score(x_train,y_train)*100 , knn.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,knn.predict(x_test)),mean_absolute_error(y_test,knn.predict(x_test)) 
```

```python
# checked above that which outof knn,randomforest and decision tree will work best for us ,which is randomforest in this case
#first check which model, than check errors ,whichever model has less error or nearest to 0 will be best model/algorithm to work n
```
